> Utilice la función creada en el punto 1 para verificar sus resultados.

**Respuesta:**

````python
def clean_data(dataframe):
    print("1- Eliminar valores nulos del dataframe")
    dataframe.replace("", np.nan, inplace=True)
    dataframe = dataframe.dropna()

    print("2- Eliminar valores fuera de rango")
    dataframe['math score'] = dataframe['math score'].astype(int)
    dataframe['reading score'] = dataframe['reading score'].astype(int)
    dataframe['writing score'] = dataframe['writing score'].astype(int)
    dataframe = dataframe[dataframe['math score'] <= 100]
    dataframe = dataframe[dataframe['reading score'] <= 100]
    dataframe = dataframe[dataframe['writing score'] <= 100]

    print("3- Eliminar registros con nombres duplicados")
    dataframe = dataframe.drop_duplicates(subset='names')

    return dataframe
````

### 3.1 Transformación de notas a "Nota Chilena" [0.3 Puntos]

En una de las reuniones con Don Caguayo este les comenta su intención de pasar todas sus notas a la "Escala Chilena". Para esto les facilita la siguiente ecuación:

$$Nota \, Chilena= \% \, de \, logro*0.06 + 1$$

Utilizando la ecuación entregada, cree una función que le permita pasar a la escala Chilena las notas del dataframe y entregue las notas redondeadas al segundo decimal.

**To-Do**

- [ ] Tranforme los valores a la escala chilena.

**Respuesta:**

````python
# Escriba su código aquí
def nota_chilena(dataframe_in, columna):
    dataframe_in['nota chilena'] = dataframe_in[columna] * 0.06 + 1 #transofmación a nota chilena
    dataframe_in['nota chilena'] = dataframe_in['nota chilena'].round(2) # redondeo a 2 decimales

    return dataframe_in
````

### 3.3 Reprobados y Destacados [0.9 Puntos, 0.3 cada DataFrame]

Contentos con su desempeño, la Universidad les solicita que obtengan los alumnos quienes han reprobado el año y los que han tenido desempeño de destacado. Para esto se considerará que un alumno ha reprobado el año cuando el promedio de notas es inferior a 4.0 en la Escala Chilena, y se considerará destacado si posee notas superiores a 5.8 en todas las asignaturas (o sea notas superiores a 5.8 en `math score`, `reading score` y `writing score`).

Cree una función que permita obtener tres Dataframes:
- El primero deberá poseer solamente los alumnos que han reprobado en la prestigiosa universidad.
- El segundo debe entregar un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados.
- El tercero será un Data Frame con los alumnos destacados ordenados de mayor a menor.

**To-Do:**

- [ ] Obtener DataFrame con los alumnos que reprobaron el año.
- [ ] Obtener DataFrame similar al original, pero marcados con una "R" el atributo GPA de los alumnos reprobados.
- [ ] Obtener DataFrame con los alumnos destacados ordenados de mayor a menor.

**Respuesta:**

````python
#Escriba su código aquí

# Muestre sus resultados aquí


def reprobados_destacados(dataframe_in):
    dataframe1 = dataframe_in[dataframe_in['promedio'] < 40] # primer df

    df_grades2 = dataframe_in.copy() #copiamos el df original para modificarlo
    df_grades2[['promedio']] = df_grades2[['promedio']].map(lambda x: 'R' if x < 40 else x) #aplicamos un mapeo reemplazando los valores bajo 4 por R
    dataframe2 = df_grades2 # segundo df

    dataframe3 = dataframe_in[dataframe_in['promedio'] > 5.8] # tercer df
    dataframe3 = dataframe3.sort_values(by='promedio', ascending=False)
    return {'DataFrame 1': dataframe1,'DataFrame 2': dataframe2, 'DataFrame 3': dataframe3}
````

        some high school -> school
        some college -> school

  Los resultados de este mapeo deben ser guardados en la columna `simple parental level of education`.

**Análisis [0.3 Puntos]**

Finalmente, responda las siguientes preguntas sobre el gráfico de caja que viene implementado a continuación. En este se muestra el GPA con respecto al nivel educacional, donde la variable de color es la etnicidad.

- [ ]  ¿Hay alguna diferencia entre los grupos graficados tanto para el nivel educacional de los padres como también para la etnicidad?


**Respuesta:**

````python
mapeo = {
    'some high school': 'school',
    'some college': 'school',
    'high school': 'school',
    'bachelor\'s degree': 'college',
    'associate\'s degree': 'college',
    'master\'s degree': 'postgraduate'
}
df_grades['simple parental level of education'] = df_grades['parental level of education'].map(mapeo)

promedio_notas = df_grades.groupby(['race/ethnicity', 'simple parental level of education'])[['math score', 'reading score', 'writing score']].mean().round(2)

numero_alumnos = df_grades.groupby(['race/ethnicity', 'simple parental level of education']).size()

total_alumnos = df_grades.shape[0]

percentage = (numero_alumnos / total_alumnos * 100).map(lambda x: f'{x:.2f}%')
numero_alumnos_df = pd.DataFrame({
    'count': numero_alumnos,
    'percentage': percentage
})

tabla = pd.merge(promedio_notas, numero_alumnos_df, on=['race/ethnicity', 'simple parental level of education'])

tabla = tabla.sort_values(by=['race/ethnicity', 'simple parental level of education'])
````

1. El outer join mantiene todas las filas de ambos dataframes. Si un alumno está en uno pero no en el otro, la información del datafreame en el que no se encuentra el alumno será NaN.

2. El left join mantiene todas las filas del dataframe de la izquierda (df_grades) y agrega la información del dataframe de la derecha (other_grades) cuando hay coincidencia.

3. El right join hace el opuesto del left join: mantiene todas las filas del dataframe de la derecha (other_grades) y agrega la información del dataframe de la izquierda (df_grades) donde hay coincidencia.

4. El inner join solo mantiene las filas donde hay coincidencia en ambos dataframes.


Para este caso, vemos como tanto el left join como el inner join dan resultado a la misma cantidad de filas. En este caso particular se produce ya que las filas de df_grades son un subonjunto de other_grades.
